```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx=pd.read_csv('/Users/kaushikchaturvedula/python/smart/Linear_X_Train.csv')
dfy=pd.read_csv('/Users/kaushikchaturvedula/python/smart/Linear_Y_Train.csv')
xtest=pd.read_csv('/Users/kaushikchaturvedula/python/smart/Linear_X_Test.csv')
x=dfx.values
y=dfy.values
x=x.reshape((-1,))
y=y.reshape((-1,))
xtest=xtest.values
xtest=xtest.reshape((-1,))
x,y,xtest
```

```python
def hypothesis(X,theta):
    return theta[0]+theta[1]*X
def error(X,Y,theta):
    err=0
    for i in range(X.shape[0]):
        hx=hypothesis(X[i],theta)
        err+=(hx-Y[i])**2
    return err
```

```python
def gradient(X,Y,theta):
    grad=np.zeros(shape=(2,))
    for i in range(X.shape[0]):
        hx=hypothesis(X[i],theta)
        grad[0]+=(hx-Y[i])
        grad[1]+=(hx-Y[i])*X[i]
    return grad
```

```python
def gradient_des(X,Y,alpha=0.000015):
    theta=np.array([2.0,0.0])
    i=0
    error_lst=[]
    theta_lst=[]
    while(i<=500):
        grad=gradient(X,Y,theta)
        e=error(X,Y,theta)
        error_lst.append(e)
        theta_lst.append((theta[0],theta[1]))
        theta[0]=theta[0]-alpha*grad[0]
        theta[1]=theta[1]-alpha*grad[1]
        i+=1
    return theta,error_lst,theta_lst
```

```python
final_theta,error_lst,theta_lst=gradient_des(x,y)
plt.scatter(x,y,color='blue',label='Training Data')
plt.plot(x,hypothesis(x,final_theta),color="red",label='prediction')
plt.legend()
plt.show()
```
